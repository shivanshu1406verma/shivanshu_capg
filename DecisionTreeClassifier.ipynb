```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, accuracy_score, r2_score, mean_absolute_error
```

```python
data=load_iris()
df=pd.DataFrame(data.data,columns=data.feature_names)
df['species']=data.target
df
```

```python
x=df.drop('species',axis=1)
y=df.species
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=DecisionTreeClassifier(max_depth=5)
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
print('Training eff. ',model.score(xtrain,ytrain))
print('Testing eff.',model.score(xtest,ytest))
```

```python
print('Precision ',precision_score(ytest,ypred,average='macro'))
print('Recall ',recall_score(ytest,ypred,average=None))
print('Accuracy  ',accuracy_score(ytest,ypred))
```
