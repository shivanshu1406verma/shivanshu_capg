```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,r2_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
```

```python
data=load_iris()
df=pd.DataFrame(data.data,columns=data.feature_names)
df['target']=data.target
df.head()
```

```python
X=df.drop('target',axis=1)
y=df['target']
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=XGBClassifier(n_estimators=100,max_depth=5,learning_rate=0.1,random_state=42)
model.fit(Xtrain,ytrain)
```

```python
ypred=model.predict(Xtest)
```

```python
model.score(Xtrain,ytrain)
```

```python
model.score(Xtest,ytest)
```

```python
r2_score(ytest,ypred)
```

```python
accuracy_score(ytest,ypred)
```

```python
precision_score(ytest,ypred,average='weighted')
```

```python
recall_score(ytest,ypred,average='weighted')
```

```python
f1_score(ytest,ypred,average='weighted')
```
