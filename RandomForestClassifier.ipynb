```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, accuracy_score, r2_score, mean_absolute_error
```

```python
cdata=load_breast_cancer()
df=pd.DataFrame(cdata.data,columns=cdata.feature_names)
df['cancer']=cdata.target
df.head()
```

```python
x=df.drop('cancer',axis=1)
y=df.cancer
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
cmod=RandomForestClassifier(n_estimators=150,max_depth=5)
cmod.fit(xtrain,ytrain)
```

```python
ycpred=cmod.predict(xtest)
ycpred
```

```python
print('Training eff. ',cmod.score(xtrain,ytrain))
print('Testing eff.',cmod.score(xtest,ytest))
```

```python
print("Accuracy:", accuracy_score(ytest, ycpred))
print("Precision:", precision_score(ytest, ycpred))
print("Recall:", recall_score(ytest, ycpred))
print("R2Score:", r2_score(ytest, ycpred))
print("MAE:", mean_absolute_error(ytest, ycpred))
```
