```python
import pandas as pd
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics  import mean_absolute_error, r2_score
```

```python
rdf=pd.read_csv('Student_Performance.csv')
rdf.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
rdf["Extracurricular Activities"]=encoder.fit_transform(rdf["Extracurricular Activities"])
rdf.head()
```

```python
x=rdf.drop('Performance Index',axis=1)
y=rdf['Performance Index']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.3,random_state=42)
```

```python
rmod=SVR(kernel='linear',C=10,epsilon=0.01)
rmod.fit(xtrain,ytrain)
```

```python
yrpred=rmod.predict(xtest)
yrpred
```

```python
print('Training eff. ',rmod.score(xtrain,ytrain))
print('Testing eff.',rmod.score(xtest,ytest))
```

```python
print('R2 Score ',r2_score(ytest,yrpred))
print('MAE ',mean_absolute_error(ytest,yrpred))
```
