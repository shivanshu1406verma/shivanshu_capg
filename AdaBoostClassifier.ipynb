```python
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics  import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
x,y=make_classification(n_samples=1000,n_features=5,n_classes=2,random_state=42)
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
basemod=DecisionTreeClassifier(max_depth=5)
```

```python
adamod=AdaBoostClassifier(estimator=basemod,n_estimators=200,random_state=42)
```

```python
adamod.fit(xtrain,ytrain)
```

```python
yadapred=adamod.predict(xtest)
```

```python
print("Accuracy:", accuracy_score(ytest, yadapred))
print("Precision:", precision_score(ytest, yadapred))
print("Recall:", recall_score(ytest, yadapred))
print("F1 Score:", f1_score(ytest, yadapred))
print("Confusion Matrix:\n", confusion_matrix(ytest, yadapred))
```
