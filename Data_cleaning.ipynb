```python
df['Property_Type'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
ordi=OrdinalEncoder(categories=[['Khordha','Cuttack','Puri','Bhubaneswar']])
#df['City'] = df['City'].astype(str) # Convert 'City' to string type
df['City']=ordi.fit_transform(df[['City']])
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Garage']=le.fit_transform(df['Garage'])
```

```python
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder(sparse_output=False)
df1 = one.fit_transform(df[['Property_Type']])
name = one.get_feature_names_out(['Property_Type'])
encod_df = pd.DataFrame(df1, columns=name)
df = pd.concat([df.drop('Property_Type', axis=1), encod_df], axis=1)
df
```

```python
df['Price'] = df['Price'].apply(convert_price)
df['Price'] = df['Price'].fillna(df['Price'].mean())

df['Area_sqft'] = pd.to_numeric(df['Area_sqft'], errors='coerce')
df['Area_sqft'] = df['Area_sqft'].fillna(df['Area_sqft'].median()).astype(float)

df['Price_Per_Sqft']=df['Price']/df['Area_sqft']
```

```python
df['House_Age']=2026-df['Year_Built']
```

```python
df
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]

outliers
```
