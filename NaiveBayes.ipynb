```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics  import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score,mean_absolute_error
```

```python
data=load_diabetes()
df=pd.DataFrame(data.data,columns=data.feature_names)
df['diabetes']=data.target
df.head()
```

```python
df['diabetes'].dtypes
```

```python
out=[]
for i in df.diabetes:
  if i>=120.0:
    out.append(1)
  else:
    out.append(0)
df['diabetes']=out
df['diabetes'].value_counts()
```

```python
df
```

```python
x = df.drop('diabetes', axis=1)
y = df['diabetes']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.6, random_state=42)
```

```python
model = GaussianNB()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
print("Confusion Matrix:", confusion_matrix(ytest, ypred))
print("Accuracy:", accuracy_score(ytest, ypred))
print("Precision:", precision_score(ytest, ypred))
print("Recall:", recall_score(ytest, ypred))
print("F1-Score:", f1_score(ytest, ypred))
```
